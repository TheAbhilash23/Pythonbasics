```python
class animals():
    def __init__(self,genus):
        self.genus = genus
class domestic(animals):
    def __init__(self,species,color,sound,genus):
        super().__init__(genus)
        self.species = species
        self.sound = sound
        self.color = color
    def AI_speaks(self):
        print("AI: We have created a domestic animal which is belongs to ",self.genus,"genus and is a commonly found breed that is",self.species,". But this particular one is rare because it is ",self.color,", and that's what makes it special")

class wild(animals):
    def __init__(self, species,favmeal,speed,sound,genus):
        super().__init__(genus)
        self.species = species
        self.sound = sound
        self.favmeal = favmeal
        self.speed = speed
    def AI_speaks(self):
        print("AI: We have created a wild animal which is looking to hunt ",self.favmeal,". With a max speed of ",self.speed,", it can hunt the",self.favmeal,"easily.")
        print("It can run upto",self.speed,"miles an hour")
    
class pet(domestic):
    def __init__(self, name, age,sex, species, color,sound, genus):
        super().__init__(species, color,sound, genus)
        self.name = name
        self.age = age
        self.sex= sex
    def attenborough(self):
        print("Attenborough: Away from the forest, is  a creature... loving, caring...with his smile he can capture your heart. It's a ",self.species,"his owners call him by",self.name," the ", self.color,"skin is one of the rarest of all its kind.")
    def Jwick(self):
        print("You killed John Wick's ",self.genus,"Did you he killed 3 guys in a bar with a pencil!")
```
